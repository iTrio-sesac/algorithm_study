```python
n,m = map(int, input().split())
tray = [list(map(int,input())) for _ in range(n)]
```

```python
tray
```

```python
visited = []

for i in range(n):
    visited_row = []
    for j in range(m):
        if tray[i][j] == 0:
            visited_row.append(False)
        else:
            visited_row.append(True)
    visited.append(visited_row)

visited
```

정답에서는 dfs(x,y)로 되어 있기 때문에 visited를 따로 함수에 변수로 안넣어주고 사용해도 문제가 될게 없다는 것을 깨달았지만
visited를 변수로 넣고 돌리면 왜 cnt가 안올라가는지는 아직 모르겠음.

```python
def dfs(x,y,visited):
    if x < 0 or x >= n or y < 0 or y >= m:
        return 0
    
    if visited[x][y] == False:
        visited[x][y] = True
        dfs(x-1,y,visited)
        dfs(x+1,y,visited)
        dfs(x,y-1,visited)
        dfs(x,y+1,visited)
        return 1
    return 0

cnt = 0
for i in range(n):
    for j in range(m):
        if dfs(i, j,visited) == 1:
            cnt += 1

print(cnt)
```

함수 변수에 visited 빼니까 잘 돌아가는 모습

```python
def dfs(x,y):
    if x < 0 or x >= n or y < 0 or y >= m:
        return 0  # tray 범위 넘어가면 0을 리턴해라
    
    if visited[x][y] == False:
        visited[x][y] = True
        dfs(x-1,y)
        dfs(x+1,y)
        dfs(x,y-1)
        dfs(x,y+1)  # 상하좌우로 한번씩 다 돌려보고
        return 1  # 1을 리턴해라
    return 0  #visited[x][y] == True일 경우엔 바로 0을 리턴하라는 뜻

cnt = 0
for i in range(n):
    for j in range(m):
        if dfs(i, j) == 1:
            cnt += 1

print(cnt)
```
