**이진 탐색:** 탐색 범위를 반으로 좁혀가며 빠르게 탐색하는 알고리즘. 

**순차 탐색 Sequential search** 이란 리스트 안에 있는 특정한 데이터를 찾기 위해 앞에서부터 데이터를 하나씩 차례대로 확인하는 방법이다. 보통 정렬되지 않은 리스트에서 데이터를 찾아야 할 때 사용한다. 

따라서 데이터의 개수가 N개일 때 최대 N번의 비교 연산이 필요하므로 순차 탐색의 최악의 경우 시간 복잡도는 O(N)이다.

### 이진 탐색: 반으로 쪼개면서 탐색하기:

**Binary search은** 배열 내부의 데이터가  정렬되어 있어야만 사용할 수  있는 알고리즘이다. 데이터 정렬되어 있다면 매우 빠르게 데이터를 찾을 수 있다는 특징이 있다.

이진 탐색은 위치를 나타내는 변수 3개를 사용하는데 탐색하고자 하는 범위의 **시작점, 끝점 & 중간점**이다. 찾으려는 데이터와 중간점 위치에 있는 데이터를 반복적으로 비교해서 원하는 데이터를 찾는게 이진 탐색 과정이다. 

- 이진 탐색은 한번 확인할 때마다 확인하는 원소의 개수가 절반씩 줄어든다는 점에서 시간 복잡도가 O(logN)이다.
- 이진 탐색 알고리즘은 한 단계를 거칠때마다 확인하는 원소가 평균적으로 절반으로 줄어든다.
- 단계마다 2로 나누는 것과 동일하므로 연산 횟수는 $log2(N)$ 에 비레한다고 할 수 있다. 이는 빅오 표기 법에 따라서 간단히 $O(logN)$ 이라고 작성한다.

```python
# Book ex. Loop (반복문)
def binary_search(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        # 찾은 경우 중간점 인덱스 반환
        # Return the index if found
        if array[mid] == target:
            return mid
        # 중간점의 값보다 찾고자 하는 값이 작은 경우 왼쪽 확인
				# If the target is smaller than the value at the midpoint,search the left side
        elif array[mid] > target:
            end = mid - 1
        # 중간점의 값보다 찾고자 하는 값이 큰 경우 오른쪽 확인
				# If the target is larger than the value at the middle point, search the right side
        else:
            start = mid + 1
    return None

# N(가게의 부품 개수) 입력 
# Input the number of parts in the store (N)
n = int(input())
# 가게에 있는 전체 부품 번호를 공백으로 구분하여 입력
# Input the entire part numbers in the store, separated by spaces
array = list(map(int, input().split()))
# 이진 탐색을 수행하기 위해 사전에 정렬 수행
# Sort the array for binary search
array.sort()

# M (손님이 확인 요청한 부품 개수)을 입력
# Input the number of parts requested by the customer (M)
m = int(input())
# 손님이 확인 요청한 전체 부품 번호를 공백으로 구분하여 입력
# Input the entire part numbers requested by the customer, separated by spaces
x = list(map(int, input().split()))

# 손님이 확인 요청한 부품 번호를 하나씩 확인
# Check each part number requested by the customer
for i in x:
    # 해당 부품이 존재하는지 확인
    # Check if the part exists
    result = binary_search(array, i, 0, n-1)
    if result != None:
        print('yes', end=' ')
    else: 
        print('no', end=' ')
```

**O(log N) Explained**

Let’s bring this all back to Big O Notation. In computer science, whenever we say O(log N), it’s actually shorthand for saying O(log2 N). We just omit that small 2 for convenience.

Recall that Big O Notation resolves the key question: if there are N data elements, how many steps will the algorithm take?

$O(log N)$ means that for N data elements, the algorithm would take *log2 N steps*. If there are 8 elements, the algorithm would take three steps, since $log2( 8) = 3.$

Said another way, if we keep dividing the 8 elements in half, it would take us three steps until we end up with 1 element.

This is *exactly* what happens with binary search. As we search for a particular item, we keep dividing the array’s cells in half until we narrow it down to the correct number.

Said simply: *O(log N) means the algorithm takes as many steps as it takes to keep halving the data elements until we remain with 1.*
