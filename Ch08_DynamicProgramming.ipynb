### 다이나믹 프로그래밍 Dynamic Programming이란?
- 다이나믹 프로그래밍은 메모리를 적절히 사용하여 수행 시간 효율성을 비약적으로 향상시키는 방법이다. 
- 이미 계산된 결과(작은 문제)는 별도의 메모리 영역에 저장하여 다시 계산하지 않도록 한다. 
- 다이나믹 프로그래밍은 동적 계획법이라고도 부른다. 이때 "동적"은 일반적인 프로그래밍 분야에서의 "동적"이 아닌 별다른 의미 없이 사용된 단어이다. 

<br>

### 다이나믹 프로그래밍 구현 방식
다이나믹 프로그래밍의 구현은 일반적으로 두 가지 방식으로 구성된다. 
1. **탑다운/하향식**
    - 재귀함수로 구현한다. 
    - **메모이제이션 방식**: 한 번 계산한 결과를 메모리 공간에 메모하는 기법. 같은 문제를 다시 호출하면 메모했던 결과를 그대로 가져온다. 값을 기록해 놓는다는 점에서 **캐싱 caching**이라고도 한다. 메모이제이션은 다이나믹 프로그래밍에 국한된 개념은 아니다. 한 번 계산된 결과를 담아 놓기만 하고 다이나믹 프로그래밍을 위해 활용하지 않을 수도 있다. 
        
2. **보텀업/상향식**   
    - 다이나믹 프로그래밍의 전형적인 형태는 보텀업 방식이다. 
    - 반복문으로 구현한다. 
    - 결과 저장용 리스트는 'DP 테이블'이라고 부른다.

<br>

### 다이나믹 프로그래밍 문제 조건
다이나믹 프로그래밍은 문제가 다음의 조건을 만족할 때 사용할 수 있다.    
1)  **최적 부분 구조 Optimal Substructure**   
큰 문제를 작은 문제로 나눌 수 있으며 작은 문제의 답을 모아서 큰 문제를 해결할 수 있다.    
2) **중복되는 부분 문제 Overlapping Subproblem**
동일한 작은 문제를 반복적으로 해결해야 한다. 

<br>

### 피보나치 수열
- 다이나믹 프로그래밍으로 해결할 수 있는 대표적인 예시로 피보나치 수열이 있다. 
- 피보나치 수열은 이전 두 항의 합을 현재의 항으로 설정하는 특징이 있는 수열이다.
- 피보나치 수열을 **점화식**(인접한 항들 사이의 관계식)으로 표현할 수 있고, 이 식을 재귀함수를 이용해 구현할 수 있다. 
- 예를 들어 어떠한 수열이 있고 시작 부분의 항에 대해서 값을 알고 있으면 모든 항에 대한 각각의 값을 구할 수 있다. 
- 단순 재귀 함수로 피보나치 수열을 해결하면 2^N 시간 복잡도를 가지게 된다. N의 값이 조금만 커지면 기하급수적으로 수행 시간이 소요된다. 
- 이러한 피보나치 수열을 다이나믹 프로그래밍의 메모이제이션으로 풀 경우 시간 복잡도는 N이 된다.  
- 프로그래밍에서는 이러한 수열을 배열이나 리스트로 표현할 수 있다. 수열 자체가 여러 개의 수가 규칙에 따라서 배열된 형태를 의미하는 것이기 때문이다. (리스트나 배열 모두 '연속된 많은 데이터를 처리한다는 점은 동일하다.)

<br>

### 다이나믹 프로그래밍 vs 분할 정복
- 다이나믹 프로그래밍과 분할 정복은 모두 **최적 부분 구조**를 가질 때 사용할 수 있다.
- 다이나믹 프로그래밍과 분할 정복의 차이점은 **부분 문제의 중복**이다.
    - 다이나믹 프로그래밍 문제에서는 각 부분 문제들이 서로 영향을 미치며 부분 문제가 중복된다.
    - 분할 정복 문제에서는 동일한 부분 문제가 반복적으로 계산되지 않는다. 

<br>

### 다이나믹 프로그래밍 문제에 접근하는 방법
- 가장 먼저 그리디, 구현, 완전 탐색 등의 아이디어로 문제를 해결할 수 있는지 검토할 수 있습니다. 
- 다른 알고리즘으로 풀이 방법이 떠오르지 않으면 다이나믹 프로그래밍을 고려해본다. 
- 일단 재귀 함수로 비효율적인 완전 탐색 프로그램을 작성한 뒤에 (탑다운) 작은 문제에서 구한 답이 큰 문제에서 그대로 사용될 수 있으면, 코드를 개선하는 방법을 사용할 수 있다.

### 예제 2. 1로 만들기 
- 정수 X가 주어졌을 때, 정수 X에 사용할 수 있는 연산은 다음과 같이 4가지다.
    1. X가 5로 나누어 떨어지면, 5로 나눈다.
    2. X가 3으로 나누어 떨어지면, 3으로 나눈다.
    3. X가 2로 나누어 떨어지면, 2로 나눈다.
    4. X에서 1을 뺀다.
- 정수 X가 주어졌을 때, 연산 4개를 적절히 사용해서 값을 1로 만들고자 한다. 연산을 사용하는 횟수의 최솟값을 출력하라. 예를 들어 정수가 26이면 다음과 같이 계싼해서 3번의 연산이 최솟값이다. 
    - 26 -> 25 -> 5 -> 1

입력 조건
- 첫째 줄에 정수 X가 주어진다. (1 <= X <= 30,000)

입력 예시   
`26`

출력 조건
- 첫째 줄에 연산을 하는 횟수의 최솟값을 출력한다.

출력 예시   
`3`

````python
# 정수 x 입력 받기
x = int(input())

# 결과 저장용 리스트인 DP 테이블 생성. x가 최대 30000까지 들어올 수 있어 리스트의 길이는 다음과 같이 설정. 
dp = [0] * 30001

# 2부터 x까지의 최적의 해를 구하는 반복문. 0과 1의 최적의 해는 따로 설정하지 않아서 초기값 0이다. 
for i in range (2, x + 1):
    # 현재 수의 최소 연산 횟수를 현재의 수에서 1을 빼는 경우의 최소 연산 횟수 + 1로 저장한다. 
    dp[i] = dp[i - 1] + 1
    # 현재의 수가 2로 나누어 떨어지는 경우
    if i % 2 == 0:
        # 현재의 수의 최소 연산 횟수는 위에서 미리 저장해둔 횟수와 현재 수를 2로 나눈 값의 최소 연산 횟수 +1 중 작은 값이 된다. 
        dp[i] = min(dp[i], dp[i // 2] + 1)
    # 현재의 수가 3으로 나누어 떨어지는 경우    
    if i % 3 == 0:
        dp[i] = min(dp[i], dp[i // 3] + 1)
    # 현재의 수가 5로 나누어 떨어지는 경우
    if i % 5 == 0:
        dp[i] = min(dp[i], dp[i // 5] + 1)

print(dp[x])
````

### 예제 3. 개미 전사
개미 전사는 부족한 식량을 충당하고자 메뚜기 마을의 식량창고를 몰래 공격하려 한다. 메뚜기 마을에는 여러 개의 식량창고가 있는데, 식량창고는 일직선으로 이어져 있다.   

각 식량창고에는 정해진 수의 식량을 저장하고 있으며, 개미 전사는 식량창고를 선택적으로 약탈하여 식량을 빼앗을 것이다. 이때 메뚜기 정찰병들은 일직선상에 존재하는 식량창고 중에서 서로 인접한 식량창고가 공격받으면 바로 알아챌 수 있다.   

따라서 개미 전사가 정찰병에게 들키지 않고 식량창고를 약탈하기 위해서는 최소 한 칸 이상 떨어진 식량창고를 약탈해야 한다.
예를 들어, 식량창고 4개가 존재한다고 가정하자. 이를 배열(리스트)로 나타내면 다음과 같다. {1, 3, 1, 5}   

이때 개미 전사는 두 번째 식량창고와 네 번째 식량창고를 선택했을 때 최댓값인 총 8개의 식량을 빼앗을 수 있다. 개미 전사는 식량창고가 이렇게 일직선 상일 때 최대한 많은 식량을 얻기를 원한다.   
개미 전사를 위해 식량창고 N개에 대한 정보가 주어졌을 때 얻을 수 있는 식량의 최댓값을 구하는 프로그램을 작성하라.

입력 조건
- 첫째 줄에 식량창고의 개수 N이 주어진다. (3 <= N <= 100)
- 둘째 줄에 공백을 기준으로 각 식량창고에 저장된 식량의 개수 K가 주어진다. (0 <= K <= 1,000)
<br>

입력 예시
```
4
1 3 1 5
```

출력 조건 
- 첫째 줄에 개미 전사가 얻을 수 있는 식량의 최댓값을 출력한다. 

출력 예시
```
8
```


````python
# N 입력 받기
n = int(input())

# 모든 식량 정보를 입력 받아 리스트 형태로 저장
array = list(map(int, input().split()))

# 보텀업 방식을 사용하기 때문에 결과 저장용 리스트인 DP 테이블 생성한다. 최대 100개 창고를 가질 수 있어서 길이가 100인 리스트를 생성한다. 
dp = [0] * 100

# 첫 번째 창고까지 얻을 수 있는 식량의 최댓값은 첫 번째 원소이므로 그대로 저장한다. 
dp[0] = array[0]

# 두 번째 창고까지 얻을 수 있는 식량의 최댓값은 첫번째 원소와 두번째 원소 중 하나가 된다.
dp[1] = max(array[0], array[1])

# 세 번째 창고부터 n번째 창고까지의 최댓값은 반복문으로 진행
for i in range(2, n):
    # i번째 창고까지의 최댓값은 i-1번째와 i-2번째 + i번째에 위치한 식량값 중 큰 값이 된다. 
    dp[i] = max(dp[i - 1], dp[i - 2] + array[i])

print(dp[n - 1])
````

### 예제 4. 바닥 공사
가로의 길이가 N, 세로의 길이가 2인 직사각형 형태의 얇은 바닥이 있다. 태일이는 이 얇은 바닥을 1 X 2의 덮개, 2 X 1의 덮개, 2 X 2의 덮개를 이용해 채우고자 한다.    

이때 바닥을 채우는 모든 경우의 수를 구하는 프로그램을 작성하시오. 예를 들어 2 X 3 크기의 바닥을 채우는 경우의 수는 5가지이다. 

입력 조건
- 첫째 줄에 N이 주어진다. (1 <= N <= 1,000)

입력 예시   
`3`

출력 조건
- 첫째 줄에 2 X N 크기의 바닥을 채우는 방법의 수를 796,796으로 나눈 나머지를 출력한다. 

출력 예시   
`5`


````python
# 정수 N 입력 받기
n = int(input())

# 결과 저장하기 위한 DP 테이블 초기화
dp = [0] * 1001

# 보텀업 방식이므로 초기값 설정
dp[1] = 1
dp[2] = 3

# 3부터 n까지의 최적의 해를 구하는 반복문
for i in range(3, n + 1):
    # 
    dp[i] = (dp[i - 1] + 2 * dp[i - 2]) % 796796

print(dp[n])
````

### 예제 5. 효율적인 화폐 구성
N가지 종류의 화폐가 있다. 이 화폐들의 개수를 최소한으로 이용해서 그 가치의 합이 M원이 되도록 하려고 한다. 이때 각 화폐는 몇 개라도 사용할 수 있으며, 사용한 화폐의 구성은 같지만 순서만 다른 것은 같은 경우로 구분한다. 예를 들어 2원, 3원 단위의 화폐가 있을 때는 15원을 만들기 위해 3원을 5개 사용하는 것이 가장 최소한의 화폐 개수이다. 

입력 조건
- 첫째 줄에 N,M이 주어진다. (1<= N 100, 1 <= M <= 10,000)
- 이후 N개의 줄에는 각 화폐의 가치가 주어진다. 화폐 가치는 10,000보다 작거나 같은 자연수이다. 

입력 예시
```
2 15
2
3
```

출력 조건
- 첫째 줄에 M원을 만들기 위한 최소한의 화폐 개수를 출력한다.
- 불가능할 때는 -1을 출력한다. 

출력 예시
```
5
```

````python
n, m = map(int, input().split())

# 화폐의 종류를 저장할 배열 생성
money = []

# n번 화페의 종류 입력 받아 배열에 저장
for _ in range(n):
    money.append(int(input()))

# 보텀업 방식의 DP 테이블 생성
dp = [10001] * (m + 1)

dp[0] = 0

# 화폐의 종류 개수만큼 DP 테이블을 갱신
for i in range(n):
    # 현재 화폐 종류 금액부터 목표 금액까지 반복
    for j in range(money[i], m + 1):
        # (i -k)원을 만드는 방법이 존재하는 경우
        if dp[j - money[i]] != 10001:
            dp[j] = min(dp[j], dp[j - money[i]] + 1)

# 만약 m원을 만드는 방법이 없는 경우
if dp[m] == 10001:
    print(-1)
else:
    print(dp[m])
````
